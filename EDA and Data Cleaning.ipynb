Scatter plot draw data points using Cartesian coordinates to show the values of numerical values. they also represent between two numerical values.


```python
# Create the data
x = [1, 3, 5, 7, 9, 11]
y = [10, 25, 35, 33, 41, 59]

# Draw the scatter chart
plt.scatter(x, y, c='blue', marker='*', alpha=0.5)

# Append the label on X-axis
plt.xlabel("X-label")

# Append the label on Y-axis
plt.ylabel("Y-label")

# Add the title to graph
plt.title("Scatter Chart Sample")

# Display the chart
plt.show()
```

Pie plot is a circular graph that is split up into wedge-shaped pieces. Each piece is proportionate to the value it represents. The total value of the pie is 100 percent

```python
# Create the data
subjects = ['Mathematics', 'Physics', 'Chemistry', 'Engineering']

scores = [85, 62, 57, 92]

# Plot the pie plot
plt.pie(scores,
        labels=subjects,
        colors=['r', 'g', 'b', 'y'],
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        explode=(0, 0.1, 0, 0),

        )

# Add title to graph
plt.title("Student Performance")

# Draw the chart
plt.show()
```

Bar Plot is a visual tool to compare the values of various groups. it can be drawn horizontally or vertically

```python
# Create the data
movie_ratings = [1, 2, 3, 4, 5]
rating_counts = [21, 45, 72, 89, 42]

# Plot the data
plt.bar(movie_ratings, rating_counts, color='blue')

# Add X label oN X-axis
plt.xlabel("Movie Rating")

# Add X label on Y-axis
plt.ylabel("Rating Frequency")

# Add a title to graph
plt.title("Movie Rating Distribution")

plt.show()
```

Histogram plot shows the distribution of a numeric variable. A histogram only works on a single variable while bar graph works on two variables:

```python
# Create the data
employee_age = [21, 28, 32, 34, 35, 35, 37, 42, 47, 55]

# Create bins for histogram
bins = [20, 30, 40, 50, 60]

# Plot the histogram
plt.hist(employee_age, bins, rwidth=0.6)

# Add X label on X-axis
plt.xlabel("Employee Age")

# Add X label on Y-axis
plt.ylabel("Frequency")

# Add title to the graph
plt.title("Employee Age Distribution")

# Show the plot
plt.show()
```

Bubble Plot is a type of scatter plot. It creates bubble on data points. Bubble shows the third dimension of a plot. It shows three numerical values: two values are on the x and y axes and the third one is the size of data points(or bubbles):

```python
import matplotlib.pyplot as plt
import numpy as np

# Set figure size
plt.figure(figsize =(8, 5))

# Create the data
countries = ['Qatar', 'Luxembourg', 'Spain', 'Chile', 'Colombia',
             'Norway', 'UAE', 'Kuwait', 'Brunei', 'Ireland']

populations = [2781682, 604245, 56668893, 78800202, 5447894,
               46749949, 68292022, 4648494, 89292020, 7779280 ]

# FIXED  must be same length as countries
gdp_per_capita = [130475, 106705, 100345, 79530, 78785,
                  74356, 69382, 67000, 60000, 65000]  # added two values

# Scale GDP per capita for bubble size
scaled_gdp_per_capita = np.divide(gdp_per_capita, 80)

colors = np.random.rand(len(countries))

# Draw the scatter diagram
plt.scatter(
    countries,
    populations,
    s=scaled_gdp_per_capita,
    c=colors,
    cmap="Blues",
    edgecolors="grey",
    alpha=0.5
)

# Add labels and title
plt.xlabel("Countries")
plt.ylabel("Population")
plt.title("Bubble Chart")

# Rotate x labels for visibility
plt.xticks(rotation=45)

# Show plot
plt.show()
```

lm plot plots the scatter and fits the regression model on it. A scatter plot is the best way to understand the relationship between two variables. Its output visualization is a joint distribution of two variables.

```python
import seaborn as sns
```

```python
df = pd.DataFrame({
    'x':[1, 3, 5, 7, 9, 11],
    'y':[10, 25, 35, 33, 41, 59]

})

# Create lmplot
sns.lmplot(x='x', y='y', data=df)

plt.show()
```

```python
# to remove the regression line

sns.lmplot(x='x', y='y', data=df, fit_reg=False)

plt.show()
```

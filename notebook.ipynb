```python
def get_city_population(populations, city):
    try:
        return populations[city]
    except KeyError:
        raise KeyError(f"City '{city}' not found in population dictionary.")


city_population = {"New York": 8400000, "London": 8900000, "Paris": 2100000}

# This will raise a KeyError with a clear message
get_city_population(city_population, "Lagos")
```

zero() : The zeros() function creates an array for a given dimension with all zeros.

ones(): The ones () function creates an array for a given dimension with all ones.

fulls(): The full() function generates an array with constant values.

eyes() : The eye() function creates an identity matrix.

random(): The random() function creates an array with any given dimension



```python
# Create an array of all zeros
p = np.zeros((3, 3))
print(p)

# Create an array of all ones
q = np.ones((2, 2))
print(q)

# Create a constant array
r = np.full((2, 2), 4)
print(r)

# Create a 2 x 2 identity matrix
s = np.eye(4)
print(s)

# Create an array filled with random values
t = np.random.random((3, 3))
print(t)
```

The stacking of NumPy arrays

Horizontal stacking: The same dimensional arrays are joined along with a horizontal axis using the hstack() and concatenate()

```python
arr1 = np.arange(1, 10).reshape(3, 3)
print(arr1)

arr2 = 2 * arr1
print(arr2)

# Horizpntal stacking
arr3 = np.hstack((arr1, arr2))
print(arr3)
```

```python
# to achieve the same horizontal stacking with concatenate with axis parameter value 1:
arr4 = np.concatenate((arr1, arr2), axis=1)
print(arr4)
```

Vertical stacking: the same dimensional arrays are joined along with a vertical axis using the vstack() and concatenate() functions

```python
arr5 = np.vstack((arr1, arr2))
print(arr5)
```

```python
arr6 = np.concatenate((arr1, arr2), axis=0)
print(arr6)
```

Depth stacking: in depth stacking, the same dimensional arrays are joined along wwith a third axis(depth) using the dstack() function

```python
arra7 = np.dstack((arr1, arr2))
print(arra7)
```

Column stacking: Column stacking stacks multiple sequence one-dimensional arrays as columns into a single two-dimensional array

```python
arr3 = np.arange(4, 7)
print(arr3)

arr4 = arr3 *  2
print(arr4)

# Create column stack
arr_col_stack = np.column_stack((arr3, arr4))
print(arr_col_stack)
```

Row stacking - stacks multiple sequence one-dimensional arrays as rows into a single two-dimensional arrays


```python
arr_row_stack = np.row_stack((arr1, arr2))
print(arr_row_stack)
```

Partitioning NumPy aarays

Horizontal spitting : the given array is divided into N equal sub-arrays along the horizontal axis using the hsplit() function

```python
# Create an array
arr = np.arange(1, 10).reshape(3, 3)
print(arr)

# horizontal splitting
arr_hor_spilt = np.hsplit(arr, 3)
print(arr_hor_spilt)
```

Vertical splitting : In vertical split, the given array is divided into N equal sub-arrays along the vertical axis using the vspilt() and the split() functions. the spilt function with axis=0 performs the same operation as the vsplit() function

```python
# vertical spilt
arr_ver_split = np.vsplit(arr, 3)
print(arr_ver_split)

# split with axis=0
arr_split = np.split(arr, 3)
print(arr_split)

# split with axis=1
arr_split2 = np.split(arr, 3, axis=1)
print(arr_split2)


```

Creating numPy views and copies

```python
arr = np.arange(1, 5).reshape(2, 2)
print(arr)

```

```python
# Create no copy nly assignment
arr_no_copy = arr

# Create Deep Copy
arr_copy = arr.copy()

# Create shallow copy using view
arr_view = arr.view()

print("Original Array:", id(arr))
print("Assignment:", id(arr_no_copy))
print("Deep Copy:", id(arr_copy))
print("Shallow Copy(View):", id(arr_view))
```

```python
# Update the values of original array
arr[1] = [99, 89]

# Check values of array view
print("View Array:\n", arr_view)
print("Copied Array:\n", arr_copy)

```

Boolean and fancy indexing : Indexing techniques help us to select and filter elements from a NumPy array. This indexing returns elements that have a true value for the Boolean expression


```python
arr =np.arange(21, 41, 2)
print("Original Array:\n", arr)

# Boolean Indexing
print("After Boolean Condition:", arr[arr>30])
```

Fancing Indexing is a special type of indexing in which elements of an array are selected by an array of indices. This means we pass the array of indices in brackets. Fancy indexing also supports multi-dimensional arrays. Ths will help us to easily select and modify a complex multi-dimensional set of arrays

```python
arr = np.arange(1, 21).reshape(5, 4)
print("Original Array:\n", arr)

# Selecting 2nd and 3rd row
indices = [0, 1]
print("Selected Ist and 2nd Row:\n", arr[indices])


# Selecting 3rd and 4th row
indices = [2, 3]
print("Selected 3rd and 4th Row:\n", arr[indices])
```

```python
row = np.array([1, 2])
col = np.array([2, 3])

print("Selected Sub-Array: ", arr[row,col])
```

Broadcasting arrays: python lists do not support direct vectorizing arithmetic operations. NumPy offers a faster vectorized array operation compared to Python list loop-based operations. Broadcasting functionality checks a set of rules for applying binary functions, such as addition, substraction, multiplication on different shapes of an array

```python
arr1 = np.arange(1, 5).reshape(2, 2)
print(arr1)
```

```python
arr2 = np.arange(5,9).reshape(2, 2)
print(arr2)
```

```python
# Add two matrices

print(arr1 + arr2)
```

```python
# Multiply two matrices
print(arr1 * arr2)
```

```python
# Add a scaler value
print(arr1 + 3)
```

```python
#Multiply with a scalar value
print(arr1 * 3)
```

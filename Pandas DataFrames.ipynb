Pandas library is designed to work with a panel or tabular data. pandas is fast, highly efficient and productive tool for manipulation and analyzing string, numeric , data and time -series data. pandas provides data structures such as DataFrames and Series. A Pandas DataFrame is a tabular, two-dimensional labeled and indexed data structure with a grid of rows and columns. Its columns are heterogenous types. It has the capability to work with different types of objects, carry out grouping and joining operations, handle missing values, create pivot tables, and deal with dates.

```python
# Import pandas library
import pandas as pd
```

```python
# Create empty DataFrame
df = pd.DataFrame()

# Header of dataframe
df.head()
```

```python
# Create dictionary of list
data = {"Name": ["Vijay", "Sundar", "Satyam", "Indira" ], "Age": [23, 45, 46, 52]}

# Create the pandas DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
data1 = [{"Name": "Vijay", "Age": 23}, {"Name": "Sundar", "Age": 45}, {"Name": "Satyam", "Age": 52}]

df = pd.DataFrame(data1)

df.head()
```

pandas Series is a one-dimensional sequential data structure that is able to handle any type of data, such as string, numeric, datatime, python lists, and dictionaries with labels and indexes. Series is one of columns of a DataFrame.

```python
# Creating Pandas Series using Dictionary
dict1 = {0 : "Ajay", 1 : "Jay", 2 : "Vijay" }

# Creat pandas Series
series = pd.Series(dict1)

# Show series
series

```

```python
import numpy as np
```

```python
arr = np.array([51, 65, 48, 59, 68])

# Create Pandas Series
series1 = pd.Series(arr)
series1
```

Using a single scalar value: To create a pandas Series with a scalar value, pass the scalar value and index list to a Series object


```python
series2 = pd.Series(10, index=[0, 1, 2, 3, 4, 5])
series2
```

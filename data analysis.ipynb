to handle duplicate values

```python
duplicates = customer_data.duplicated()
print(duplicates)

duplicate_rows = customer_data[customer_data.duplicated()]
print(duplicate_rows)

customer_data = customer_data.drop_duplicates()

```

```python

# Keep the last occurrence of each duplicate
customer_data = customer_data.drop_duplicates(keep="last")

# Keep the first occurrence of each duplicate
customer_data = customer_data.drop_duplicates(keep="first")

# Remove all duplicate rows
customer_data = customer_data.drop_duplicates(keep=False)
```
